# Model specification

## Parameters to be estimated

```python
ASC_EV = Beta('ASC_EV', 0, None, None, 0)
ASC_ICE = Beta('ASC_ICE', 0, None, None, 0)
B_AGE_EV = Beta('B_AGE_EV', 0, None, None, 0)
B_AGE_ICE = Beta('B_AGE_ICE', 0, None, None, 0)
B_GENDER_EV = Beta ('B_GENDER_EV', 0, None, None, 0)
B_GENDER_ICE = Beta ('B_GENDER_ICE', 0, None, None, 0)
B_INCOME_EV = Beta('B_INCOME_EV', 0, None, None, 0)
B_INCOME_ICE = Beta('B_INCOME_ICE', 0, None, None, 0)
B_PEOPLE_EV = Beta('B_PEOPLE_EV', 0, None, None, 0)
B_PEOPLE_ICE = Beta('B_PEOPLE_ICE', 0, None, None, 0)
B_EQ_EV = Beta('B_EQ_EV', 0, None, None, 0)
B_EQ_ICE = Beta('B_EQ_ICE', 0, None, None, 0)
B_OCCUP_STU_EV = Beta('B_OCCUP_STU_EV', 0, None, None, 0)
B_OCCUP_STU_ICE = Beta('B_OCCUP_STU_ICE', 0, None, None, 0)
B_OCCUP_JOB_EV = Beta('B_OCCUP_JOB_EV', 0, None, None, 0)
B_OCCUP_JOB_ICE = Beta('B_OCCUP_JOB_ICE', 0, None, None, 0)
B_V_OWN_EV = Beta('B_V_OWN_EV', 0, None, None, 0)
B_V_OWN_ICE = Beta('B_V_OWN_ICE', 0, None, None, 0)
B_PURCHASE_COST_EV = Beta('B_PURCHASE_COST_EV', 0, None, None, 0)
B_PURCHASE_COST_ICE = Beta('B_PURCHASE_COST_ICE', 0, None, None, 0)
B_OPERATING_COST_EV = Beta('B_OPERATING_COST_EV', 0, None, None, 0)
B_OPERATING_COST_ICE = Beta('B_OPERATING_COST_ICE', 0, None, None, 0)
B_MAINTENANCE_COST_EV = Beta('B_MAINTENANCE_COST_EV', 0, None, None, 0)
B_MAINTENANCE_COST_ICE = Beta('B_MAINTENANCE_COST_ICE', 0, None, None, 0)
B_RESALE_VALUE_EV = Beta('B_RESALE_VALUE_EV', 0, None, None, 0)
B_RESALE_VALUE_ICE = Beta('B_RESALE_VALUE_ICE', 0, None, None, 0)
B_DRIVING_RANGE_EV = Beta('B_DRIVING_RANGE_EV', 0, None, None, 0)
B_DRIVING_RANGE_ICE = Beta('B_DRIVING_RANGE_ICE', 0, None, None, 0)
B_AVAIL_CHAGE_FUEL_EV = Beta('B_AVAIL_CHAGE_FUEL_EV', 0, None, None, 0)
B_AVAIL_CHAGE_FUEL_ICE = Beta('B_AVAIL_CHAGE_FUEL_ICE', 0, None, None, 0)
B_QUICK_ACCELERATION_EV = Beta('B_QUICK_ACCELERATION_EV', 0, None, None, 0)
B_QUICK_ACCELERATION_ICE = Beta('B_QUICK_ACCELERATION_ICE', 0, None, None, 0)
B_SMOOTH_RIDING_EXPERIENCE_EV = Beta('B_SMOOTH_RIDING_EXPERIENCE_EV', 0, None, None, 0)
B_SMOOTH_RIDING_EXPERIENCE_ICE = Beta('B_SMOOTH_RIDING_EXPERIENCE_ICE', 0, None, None, 0)
B_INTERIOR_SPACE_EV = Beta('B_INTERIOR_SPACE_EV', 0, None, None, 0)
B_INTERIOR_SPACE_ICE = Beta('B_INTERIOR_SPACE_ICE', 0, None, None, 0)
B_NOISE_LEVEL_EV = Beta('B_NOISE_LEVEL_EV', 0, None, None, 0)
B_NOISE_LEVEL_ICE = Beta('B_NOISE_LEVEL_ICE', 0, None, None, 0)
B_ENVIRONMENTAL_PROTECTION_EV = Beta('B_ENVIRONMENTAL_PROTECTION_EV', 0, None, None, 0)
B_ENVIRONMENTAL_PROTECTION_ICE = Beta('B_ENVIRONMENTAL_PROTECTION_ICE', 0, None, None, 0)
B_HEAR_EV_EV = Beta('B_HEAR_EV_EV', 0, None, None, 0)
B_HEAR_EV_ICE = Beta('B_HEAR_EV_ICE', 0, None, None, 0)
B_KNOW_EV_EV = Beta('B_KNOW_EV_EV', 0, None, None, 0)
B_KNOW_EV_ICE = Beta('B_KNOW_EV_ICE', 0, None, None, 0)
B_COST_AWARE_EV_EV = Beta('B_COST_AWARE_EV_EV', 0, None, None, 0)
B_COST_AWARE_EV_ICE = Beta('B_COST_AWARE_EV_ICE', 0, None, None, 0)
B_SUSTAIN_EV_EV = Beta('B_SUSTAIN_EV_EV', 0, None, None, 0)
B_SUSTAIN_EV_ICE = Beta('B_SUSTAIN_EV_ICE', 0, None, None, 0)
```

#### Specification of the utility functions

```python
V1 = ASC_EV + \
     B_AGE_EV * AGE + \
     B_GENDER_EV * GENDER + \
     B_INCOME_EV * INCOME + \
     B_PEOPLE_EV * PEOPLE +\
     B_EQ_EV * EQ + \
     B_OCCUP_STU_EV * OCCUP_STU + \
     B_OCCUP_JOB_EV * OCCUP_JOB + \
     B_V_OWN_EV * V_OWN + \
     B_PURCHASE_COST_EV * PURCHASE_COST + \
     B_OPERATING_COST_EV * OPERATING_COST + \
     B_MAINTENANCE_COST_EV * MAINTENANCE_COST + \
     B_RESALE_VALUE_EV * RESALE_VALUE + \
     B_DRIVING_RANGE_EV* DRIVING_RANGE + \
     B_AVAIL_CHAGE_FUEL_EV * AVAIL_CHAGE_FUEL + \
     B_QUICK_ACCELERATION_EV * QUICK_ACCELERATION + \
     B_SMOOTH_RIDING_EXPERIENCE_EV * SMOOTH_RIDING_EXPERIENCE + \
     B_INTERIOR_SPACE_EV * INTERIOR_SPACE + \
     B_NOISE_LEVEL_EV * NOISE_LEVEL + \
     B_ENVIRONMENTAL_PROTECTION_EV * ENVIRONMENTAL_PROTECTION + \
     B_HEAR_EV_EV * HEAR_EV + \
     B_KNOW_EV_EV * KNOW_EV + \
     B_COST_AWARE_EV_EV * COST_AWARE_EV + \
     B_SUSTAIN_EV_EV * SUSTAIN_EV

V2 = ASC_ICE + \
     B_AGE_ICE * AGE + \
     B_GENDER_ICE * GENDER + \
     B_INCOME_ICE * INCOME + \
     B_PEOPLE_ICE * PEOPLE +\
     B_EQ_ICE * EQ + \
     B_OCCUP_STU_ICE * OCCUP_STU + \
     B_OCCUP_JOB_ICE * OCCUP_JOB + \
     B_V_OWN_ICE * V_OWN + \
     B_PURCHASE_COST_ICE * PURCHASE_COST + \
     B_OPERATING_COST_ICE * OPERATING_COST + \
     B_MAINTENANCE_COST_ICE * MAINTENANCE_COST + \
     B_RESALE_VALUE_ICE * RESALE_VALUE + \
     B_DRIVING_RANGE_ICE * DRIVING_RANGE + \
     B_AVAIL_CHAGE_FUEL_ICE * AVAIL_CHAGE_FUEL + \
     B_QUICK_ACCELERATION_ICE * QUICK_ACCELERATION + \
     B_SMOOTH_RIDING_EXPERIENCE_ICE * SMOOTH_RIDING_EXPERIENCE + \
     B_INTERIOR_SPACE_ICE * INTERIOR_SPACE + \
     B_NOISE_LEVEL_ICE * NOISE_LEVEL + \
     B_ENVIRONMENTAL_PROTECTION_ICE * ENVIRONMENTAL_PROTECTION + \
     B_HEAR_EV_ICE * HEAR_EV + \
     B_KNOW_EV_ICE * KNOW_EV + \
     B_COST_AWARE_EV_ICE * COST_AWARE_EV + \
     B_SUSTAIN_EV_ICE * SUSTAIN_EV
```

#### Associate the utility functions with the numbering of the alternatives

```python
V = {1: V1 ,
     2: V2}
```

#### Associate the availability conditions with the alternatives

```python
av = {1: EV_AV ,
      2: ICE_AV}
```

```python
logprob = models.loglogit (V, av, CHOICE)
```

#### Biogeme

```python
biogeme = bio.BIOGEME(database, logprob)
biogeme.modelName = 'Thesis Final'
```
